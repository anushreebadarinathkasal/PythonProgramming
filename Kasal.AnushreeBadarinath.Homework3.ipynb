## Solution1
###  __Input:__ 
#### *user enters the filename to be used*
### __Expected Output:__  
#### *The program picks a random word and scrambles it. It asks user to guess the answer or meaning of the word.*
### __Algorithm:__
#### *1. Random word is picked.*
#### *2. The picked word is scrambled.*
#### *3. The program asks to guess the word or for meaning of the word.*
#### *4. Displays whether the user guessed it correctly or not.*
### __Instructions to Execute:__ 
#### *Click on the below cell and execute run button.*

```python
import random
filename = input("Enter the file name:")         #Enter the filename
try:                                     
    fileopen=open(filename,'r')                  #Read the file
    words_meanings ={}                           #Create empty dictionary
    flag =0
    for line in fileopen:                        #Parse the file line by line
        k,v = line.strip().split(',')            #split by comma word and its meaning
        words_meanings[k.strip()]=v.strip()      #add the word as key and meaning as value to the dictionary
   
    word = random.choice(list(words_meanings.keys()))    #pick a random word
    correct_word = word                               
    correct_meaning = words_meanings[word]
    l = list(word)
    random.shuffle(l)                                    #Logic to shuffle the letters in the word                            
    jumble = ''.join(l)                                  #form the jumbled word
   
    print("Unscramble the following letters to form a word. Type “?” for the meaning of the unscrambled word:",jumble)
    while True:
        userin = input("Enter the answer [or ? for the meaning]:") #The program promting the user to enter word or ask meaning
        if(userin == "?"):                                         #If ? display meaning of word
            print("The word means:",correct_meaning)
        else:
            if(userin.casefold() == correct_word.casefold()):      #Compare user input with correct word
                break
            else:                                                  #if the comparison is incorrect display message
                print("Wrong, try again")
           
           

   
    while True:
        if flag == 1:
            flag = 0
            val = input("Do you want to continue yes or no:")#display message the user guessed it correct and ask to continue or not
        else:
            val = input("You Got it!.Do you want to continue yes or no:")#display message the user guessed it correct and ask to continue or not
        
        if(val.casefold() == "yes"):                                 
            word = random.choice(list(words_meanings.keys()))        #pick a random word
            correct_word = word
            correct_meaning = words_meanings[word]
            l = list(word)
            random.shuffle(l)                                        #Logic to shuffle the letters in the word
            jumble = ''.join(l)                                      #form the jumbled word
   
            print("Unscramble the following letters to form a word. Type “?” for the meaning of the unscrambled word:",jumble)
            while True:
                userin = input("Enter the answer [or ? for the meaning]:")#The program promting the user to enter word or ask meaning
                if(userin == "?"):                                        #If ? display meaning of word
                    print("The word means:",correct_meaning)
                else:
                    if(userin.casefold() == correct_word.casefold()):   #Compare user input with correct word
                        break
                    else:
                        print("Wrong, try again")                       #if the comparison is incorrect display message
        elif(val.casefold() == "no"):  
            print("Goodbye!")                                           #display exit message
            break
        else:
            print("Invalid input , type yes or no")                                           #display exit message
            flag = 1
   
    fileopen.close()
except:
    print("file not found")                                             #If file is not found print message
```

## Solution2
###  __Input:__ 
#### *user enters the filename to be used*
#### *user enters option 1, 2 and 3 as prompted by program*
#### *user enters reviewers name*
### __Expected Output:__  
#### *The program promts the user to select 1 , 2 and 3 option to calculate the similarities score using euclidean distances*
### __Algorithm:__
#### *1. Checks if the reviewers name exists in the dictionary.*
#### *2. Consider only the restaurants that are reviewed by both reviewers .*
#### *3. Calculate the similarities score.*
#### *4. Displays the score.*
### __Instructions to Execute:__ 
#### *Click on the below cell and execute run button.*

```python
filename = input("Give the name of the restaurant reviews file:")   #Enter the filename
try:
    file = open(filename,"r")                                       #read the file
    dict={}                                                         #create empty dictionary
    for line in file:
        list = line.rstrip().split(',')                             #removing \n and spliiting each line based on ','
        if(list[0] != ''):                                          
            dict[list[0]] = {}                                      #first string as key as name of reviewer
            for i in range(len(list)): 
                if((i % 2 != 0) and (i != 0)):
                    dict[list[0]][list[i]] = float(list[i+1])      #add the following restaurant and review scores for each reviewer
    while(1):
        option = input("What do you want to do? Input 1 for similarity between two reviewers, or Input 2 for similarity between one reviewer and all others in the database or 3 to quit:")
        try:
            option = int(option)                                       #check if the number entered is integer
        except ValueError:
            option = -1
        if (option == 1):                                              
            reviewer1 = input("Provide Reviewer1 name:")              #enter reviewer 1
            reviewer2 = input("Provide Reviewer2 name:")              #enter reviewer 2
            
            if reviewer1 in dict.keys() and reviewer2 in dict.keys():  #logic to check if reviewers name exists in dictionary
                
                sum = 0.0;
                for key1 in (dict[reviewer1].keys()):               
                    if(key1 in dict[reviewer2].keys()):             # if the key exists in reviewer2 only then consider
                        sum = sum + float( pow((dict[reviewer1][key1]-dict[reviewer2][key1]),2)) #compute the score

                result = pow(sum,0.5)                              # calculate euclidean distance
                print("The similarity score between "+reviewer1+" and "+reviewer2+" is:"+str(result)) # print message of scores
            else:
                print("The entered reviewer name does not exist")  #print message if reviewers does not exist

        elif (option == 2):
            reviewer = input("Provide Reviewer name:")             #enter reviewer name
            
            if reviewer in dict.keys():                            #logic to check if reviewer exists in dictionary
                print("The Similarity Scores are:")                
                for key1 in (dict.keys()):                         
                    sum = 0.0
                    if(reviewer != key1):                          #if its not the reviewer then proceed
                        for key2 in (dict[reviewer].keys()):       #for each keys in reviewer iterate
                            if(key2 in dict[key1].keys()):         #compute the score 
                                sum = sum + float(pow((dict[key1][key2]-dict[reviewer][key2]),2))
                        result = round(pow(sum,0.5),2)             #calculate euclidean distance
                        print((reviewer.ljust(20)+key1.ljust(20)+str(result).ljust(20)))
            else:
                print("The entered reviewer name does not exist") #print message if reviewers does not exist

        elif (option == 3):                                       #Quit if option 3
            print("Goodbye!")
            break
        else:
            print("INVALID OPTION!!!")

except:                                                           #If file is not found print message
    print("file not found")
```
