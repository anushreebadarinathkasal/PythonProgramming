## Solution1
###  __Input:__ 
#### *seven faced dice is rolled hundred times*
### __Expected Output:__  
#### *The program should display Number of 1's , 2's, 3's, 4's, 5's, 6's ,7's and also graphical representation with asterisk.*
### __Algorithm:__
#### *1. Seven faced dice is rolled.*
#### *2. Counters are incremented according to the value of the dice.*
#### *3. Display the final results after hundred rolls.*
#### *4. Display the results in graphical representation.*
### __Instructions to Execute:__ 
#### *Click on the below cell and execute run button.*

```python
import random                                          #Importing random to implement randint()
i=0                                                    #counters and variables initialized to 0
numberofones=0
numberoftwos=0
numberofthrees=0
numberoffours=0
numberoffives=0
numberofsixs=0
numberofsevens=0                                       
while i<100:                                          #while loop for 100 times to rolll the dice
    rolled_num = random.randint(1,7);                 #random number generator between 1 and 7
    if(rolled_num ==1):                               #check to compare rolled value is 1
        numberofones += 1                             #Increment the numberofones counter
    elif(rolled_num ==2):                             #check to compare rolled value is 2
        numberoftwos += 1                             #Increment the numberoftwos counter
    elif(rolled_num ==3):                             #check to compare rolled value is 3
        numberofthrees += 1                           #Increment the numberofthrees counter
    elif(rolled_num ==4):                             #check to compare rolled value is 4
        numberoffours += 1                            #Increment the numberoffours counter
    elif(rolled_num ==5):                             #check to compare rolled value is 5
        numberoffives += 1                            #Increment the numberoffives counter
    elif(rolled_num ==6):                             #check to compare rolled value is 6
        numberofsixs += 1                             #Increment the numberofsixs counter
    else:
        numberofsevens +=1                            #Increment the numberofsevens counter
        
    i +=1                                             #Increment the counter to keep track of number of rolls
    
print("Frequencies:")                         
print("Number of 1s:",numberofones)                   #display the results of all counters
print("Number of 2s:",numberoftwos)
print("Number of 3s:",numberofthrees)
print("Number of 4s:",numberoffours)
print("Number of 5s:",numberoffives)
print("Number of 6s:",numberofsixs)
print("Number of 7s:",numberofsevens)

for x in range(numberofones):                         #Print number of ones with asterisk graphical represenation 
    
    if(x==0):
        print("1    *", end = '')                     #argument end is used in print statement to avoid going to new line
    elif(x!=numberofones-1):
        print("*",end = '')
    else:
        print("*")

for x in range(numberoftwos):                         #Print number of twos with asterisk graphical represenation
    
    if(x==0):
        print("2    *", end = '')
    elif(x!=numberoftwos-1):
        print("*",end = '')
    else:
        print("*")
        
for x in range(numberofthrees):                       #Print number of threes with asterisk graphical represenation
    
    if(x==0):
        print("3    *", end = '')
    elif(x!=numberofthrees-1):
        print("*",end = '')
    else:
        print("*")
        
for x in range(numberoffours):                        #Print number of fours with asterisk graphical represenation
    
    if(x==0):
        print("4    *", end = '')
    elif(x!=numberoffours-1):
        print("*",end = '')
    else:
        print("*")

for x in range(numberoffives):                        #Print number of fives with asterisk graphical represenation
    
    if(x==0):
        print("5    *", end = '')
    elif(x!=numberoffives-1):
        print("*",end = '')
    else:
        print("*")
        
for x in range(numberofsixs):                         #Print number of sixs with asterisk graphical represenation
    
    if(x==0):
        print("6    *", end = '')
    elif(x!=numberofsixs-1):
        print("*",end = '')
    else:
        print("*")

for x in range(numberofsevens):                       #Print number of sevens with asterisk graphical represenation
    
    if(x==0):
        print("7    *", end = '')
    elif(x!=numberofsevens-1):
        print("*",end = '')
    else:
        print("*")
    
```

## Solution2
###  __Input:__ 
#### *User enters a number between 1 and 100 inclusive*
### __Expected Output:__  
#### *The program should display the number of guesses it took for computer to guess the input number entered by user*
####  *Displays error message if number entered is out of range*
### __Algorithm:__
#### *1. User should enter a number between 1 and 100 inclusive.*
#### *2. Generate and random number within 1 to 100 range and do equality check with the user entered num.*
#### *3. Repeat the random num generation and equality check till the computer guesses the correct user num.*
#### *4. Display the number of guesses it took for computer to guess the user entered num.*
### __Instructions to Execute:__ 
#### *Click on the below cell and execute run button and enter the number within range.*


```python
import random
entered_num = input("Enter a num in between 1 and 100 inclusive: ")       #prompt user to enter a number withing range


try:
    entered_num = int(entered_num)                                       #check if the number entered is integer
except ValueError:
    entered_num = -1

if entered_num <=0 or entered_num >100:                                  #check if the number entered number is within range                                
    print("Please Enter number in range of 1 to 100 inclusive")          #display error message to enter number within range
else:
    
    program_gen_value = random.randint(1,100);                           # generate random number between 1 and 100
    no_of_guesses =1                                           
    while True:
        if(program_gen_value == entered_num):                           #equality check to compare numbers
            print("You Entered {} , and it took program {} guesses".format(entered_num,no_of_guesses))
            break
        else:
            no_of_guesses +=1                                           #increment the number of guesses
            program_gen_value = random.randint(1,100);                  #generate random number again

        
    

    
```

## Solution3
###  __Input:__ 
#### *User enters a line of text*
#### *User enters a single letter to use*
### __Expected Output:__  
#### *The program should display how many times the letter entered by user occurs as first letter in the text*
####  *Displays error message if user does not enter a single letter*
### __Algorithm:__
#### *1. User should enter a line of text.*
#### *2. User Should enter a single letter*
#### *3. Program checks for the how many times the single letter entered by user occurs as first letter in the words.*
#### *4. Display the number of times the letter occurs as first letter in the text entered.*
#### *5. Display Error message if user does not enter single letter.* 
### __Instructions to Execute:__ 
#### *Click on the below cell and execute run button .Enter a line of text and a single letter to use.*


```python
text_entered = input("enter a line of text:")                       #prompts the user to enter a line of text
letter_to_use = input("enter a letter to use:")                     #Prompts the user to enter a single letter
no_of_times =0                                                      #Initialize to 0 for no_of_times counter
# need to check if user entered only a single letter
if(len(letter_to_use) ==1 ):                                        #check if user entered single letter
    for word in text_entered.split():                               #split the text enetered into words and loop through it
        if(letter_to_use.casefold() == word[0:1].casefold()):       #Comparsion to check if the entered letter is equal to first letter of a word
            no_of_times +=1                                         #Increment the counter if the comparsion check is true

    print("your letter {} occurs as first letter : {} times".format(letter_to_use,no_of_times)) #Display message with final results
else:
    print("Please enter a single letter")                           # Display message to enter a single letter
        
    
```

## Solution4
###  __Input:__ 
#### *User enters a string*
### __Expected Output:__  
#### *The program should display if the given string is pangram or displays the letters that are missing if it is not pangram*
### __Algorithm:__
#### *1. User should enter a string.*
#### *2. Program checks if the given string is pangram and displays if the string is pangram*
#### *3. Display the missing letters if it is not pangram.* 
### __Instructions to Execute:__ 
#### *Click on the below cell and execute run button.Once the function definition is executed click the next cell and execute run* 
#### *Enter a String.*

```python
def is_pangram(input_str):                                           #function definition for is pangram 
    all_alphabets = "abcdefghijklmnopqrstuvwxyz"                     #all the alphabets in english
    missing_letters = []                                                   
    for i in all_alphabets:                                          #Loop through all the alphabets
        if i not in input_str.lower():                               #check if the alphabet exists in the input string
            missing_letters.append(i)                                #If alphabets are missing add to missing letters to keep track
    
    if len(missing_letters) !=0:                                     #If there are missing letters print them out
        print("The Given String is not a pangram, you are missing letters:", end='')
        print(missing_letters)
    else:                                                            #If there are no missing letters print message saying the given string is pangram
        print("The given string is pangram")
        
    
        
```

```python
input_str = input("enter a string:")
is_pangram(input_str)
```

## Solution5
###  __Input:__ 
#### *User enters the filename to be used*
### __Expected Output:__  
#### *The program should display num of lines,num of words,number of characters and average length of a word in the file*
#### *Display Error message if the file does not exist*
### __Algorithm:__
#### *1. Enter the filename.*
#### *2. Program reads the file and displays the num of lines,num of wors,num of characters and average length of a word.*
#### *3. Display if the file does not exist* 
### __Instructions to Execute:__ 
#### *Click on the below cell and execute run button.Enter the filename to be used*

```python
filename=input("Enter the file name:")                            #Enter the filename to be used
num_of_lines =0                                                   #Initialize parameters to zero
num_of_words=0
num_of_characters=0
sum_of_word_len=0
avg_len_of_word =0

try:
    fileopen=open(filename,'r')                                   #Filehandle to user entered filename
    for line in fileopen:                                         #Read the file line by line
        words=line.split()                                        #Split the line to get words int he given line
        for eachword in words:                                    #Loop through each word
            sum_of_word_len += len(eachword)                      #Add length of each word 
        num_of_lines +=1                                          #Increment num of lines counter
        num_of_words += len(words)                                #Increment number of words counter
    
    num_of_characters = sum_of_word_len                           #Calculate total number of characters
    avg_len_of_word =sum_of_word_len/num_of_words                 #Calculate avg length of each word
        
        
    print("Number of lines are:",num_of_lines)                    #Display all the parameters calculated
    print("Number of words are:",num_of_words)
    print("Number of characters are:",num_of_characters)
    print("Average length of each word is:",avg_len_of_word)
    fileopen.close()                                              #Close the handle
            
except:
    print("File not found")                                       #Display message if file not found
    fileopen.close()
```
