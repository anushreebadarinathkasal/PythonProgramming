## Solution3
###  __Input:__ 
#### *positive integer*
### __Expected Output:__  
#### *The program should display if the number is prime or composite*
#### *Displays error message if the user does not enter integer values or positive integer number*
### __Algorithm:__
####  *1. User should enter positive  integer*
####  *2. Determines entered integer is prime or composite*
#### *3. Displays whether the positive integer is prime or composite*
### __Instructions to Execute:__ 
####  *Click on the below cell and execute run button.Enter positive integer value as prompted.*

```python
a = input("Enter positive integer")   #Assiging the user inputted string to variable a
# Error Checking to make sure that the enterd value is integer
try:
    val = int(a)  
except ValueError:
    val = -1
if val < 0:                          #Logic to check if the entered value is positive integer
    print("You dint enter a positive integer") #Displays error message
else:
    if val == 0 or val ==1:
        print("Number is neither prime nor composite") #Display this message if user enters either 0 or 1
    else:
        for i in range(2,val):                    #Logic to determine whether a number is prime or not
            if (val % i)== 0:
                print("It is a composite number")   #display if the entered positive integer is a composite number
                break
        else:
                print("It is a prime number")       #displays if the entered positive integer is a prime number
    
```

## Solution4
###  __Input:__ 
#### *Principle amount, interest rate and years*
### __Expected Output:__  
#### *The program should calculate simple interest and display*
#### *Displays error message if the user enters invalid input*
### __Algorithm:__
#### *1. User should enter  principle amount, interest rate and years*
#### *2. Determines simple interest.*
#### *3. Displays simple interest*
### __Instructions to Execute:__ 
#### *Click on the below cell and execute run button.Once the function definition is executed click the next cell and execute run* 
#### *Enter Principle amount, interest rate and years as prompted.*

```python
def SimpleInterest(principle_amount,interest_rate,years):     #function definition for simple interest
    return ((principle_amount * years * interest_rate) / 100)  #Calculation of simple interest
```

```python
principle_amount = input("Enter a Principle amount : ")   #Assiging the user inputted string to variable principle_amount
interest_rate = input("Enter Interest rate : ")        #Assiging the user inputted string to variable interest_rate
years = input("Enter number of years : ")     #Assiging the user inputted string to variable years
# Error Checking to make sure that the enterd value is number
try:
    principle_amount = float(principle_amount)
except ValueError:
    principle_amount = "error"
    print("enter correct principle amount")
    
try:
    interest_rate  = float(interest_rate)
except ValueError:
    interest_rate = "error"
    print("enter correct interest rate")

try:
    years  = float(years)
except ValueError:
    years = "error"
    print("enter correct years")

    #If user entered strings insted of numbers dont call simple interest function else execute the function call and display message accordingly
if principle_amount != "error" and interest_rate != "error" and years != "error":
                    
    if (float(principle_amount) >=0 )and (float(interest_rate)>=0)and(int(years)>0): #Check to validate numbers are positive
        si = SimpleInterest(float(principle_amount),float(interest_rate),int(years))  #simple interest function call
        print("The simple interest is : " , si) #display the simple interest value
    else:
        print("Please enter positive numbers")  #display if the number eneterd is negative
else:
    print("Please enter numbers")    #display if the user inputs other than numbers
```

### __Algorithm:__

```python
def InsideQuadrilateral():                #function definition of InsideQuadrilateral
    x1 = input("Enter x1 : ")     #Assiging the user inputted string to variable x1
    y1 = input("Enter y1 : ")     #Assiging the user inputted string to variable y1
    x2 = input("Enter x2 : ")     #Assiging the user inputted string to variable x2
    y2 = input("Enter y2 : ")     #Assiging the user inputted string to variable y2
    x = input("Enter x : ")       #Assiging the user inputted string to variable x
    y = input("Enter y : ")       #Assiging the user inputted string to variable y
    try:
        x1 = float(x1)
        x2 = float(x2)
        y1 = float(y1)
        y2 = float(y2)
        x = float(x)
        y = float(y)

        print (((x2<=x and x<=x1)or(x1<=x and x<=x2))and((y2<=y and y<=y1)or(y1<=y and y<=y2)))
          #determine if coordinates x,y lies within x1,y1 and x2,y2
    except:
        print("Please enter the correct input")  #display if the entered values are not numbers

```

```python
InsideQuadrilateral()                  #function call to InsideQuadrilateral function
while True:                            #loop to check if a user wants to continue or not
    val = input("Do you want to continue : ")
    if val == "Y" or val =='y':       #continue function call if user inputs y
        InsideQuadrilateral()
    else:
        print("GoodBye!")             #display exit message and quit
        break
        
    
```
